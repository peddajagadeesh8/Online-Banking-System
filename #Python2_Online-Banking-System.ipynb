<h2>Step 6. Data Visualization </h2>

```python
# Query data from the 'Customers' table and count the number of customers in each age group
cursor.execute('SELECT CASE \
                        WHEN Age >= 18 AND Age <= 25 THEN "18-25" \
                        WHEN Age >= 26 AND Age <= 35 THEN "26-35" \
                        WHEN Age >= 36 AND Age <= 45 THEN "36-45" \
                        WHEN Age >= 46 AND Age <= 55 THEN "46-55" \
                        ELSE "56+" \
                    END AS AgeGroup, \
                    COUNT(*) AS CustomerCount \
                FROM Customers \
                GROUP BY AgeGroup;')

# Fetch the data and store it into a Pandas DataFrame
customerCount_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Plot a bar chart of the number of customers per age group
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.bar(customerCount_data['AgeGroup'], customerCount_data['CustomerCount'], color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Customer Count')
plt.title('Number of Customers per Age Group')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Query data from the 'Accounts' table and count the number of accounts of each type
cursor.execute('SELECT AccountType, COUNT(*) AS AccountCount FROM Accounts GROUP BY AccountType;')

# Fetch the data and store it into a Pandas DataFrame
accountCount_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Plot a pie chart of the distribution of account types
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))
plt.pie(accountCount_data['AccountCount'], labels=accountCount_data['AccountType'], autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Account Types')
plt.tight_layout()
plt.show()
```

```python
# Query data from the 'Transactions' table
cursor.execute('SELECT TransactionID, AccountID, TransactionType, Amount, TransactionDate FROM Transactions')

# Fetch the data and store it into a Pandas DataFrame
transactions_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Plot a scatter plot of the relationship between 'Amount' and 'TransactionDate'
plt.figure(figsize=(12, 6))
plt.scatter(transactions_df['TransactionDate'], transactions_df['Amount'], color='skyblue')
plt.xlabel('Transaction Date')
plt.ylabel('Amount')
plt.title('Relationship between Amount and Transaction Date')
plt.tight_layout()
plt.show()
```

```python
# Query data from the 'Card' table
cursor.execute('SELECT CardID, CardNumber, CardHolderName, ExpiryDate, CVV, CardType, IssuingBank, CustomerID FROM Card')

# Fetch the data and store it into a Pandas DataFrame
card_df = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Plot a bar chart for the 'CardType' column
plt.figure(figsize=(12, 6))
card_df['CardType'].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel('Card Type')
plt.ylabel('Count')
plt.title('Card Type Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Query data from the 'Loans' and 'Customers' tables and join them on the 'CustomerID' column
cursor.execute('SELECT Loans.LoanAmount, Loans.InterestRate, Customers.Age FROM Loans \
                JOIN Customers ON Loans.CustomerID = Customers.CustomerID;')

# Fetch the data and store it into a Pandas DataFrame
loan_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Calculate the correlation matrix
corr_matrix = loan_data.corr()

# Plot a heatmap of the correlation matrix
import seaborn as sns
plt.figure(figsize=(8, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Loan Amount, Interest Rate, and Customer Age')
plt.tight_layout()
plt.show()
```

```python
# Assuming 'TransactionDate' is a datetime column
transactions_df['Day'] = transactions_df['TransactionDate'].dt.day_name()
transactions_df['Hour'] = transactions_df['TransactionDate'].dt.hour

# Pivot table for heatmap
heatmap_data = transactions_df.pivot_table(index='Day', columns='Hour', values='Amount', aggfunc='sum')

# Set the figure size for the heatmap
plt.figure(figsize=(12, 8))

# Create a heatmap using the heatmap_data DataFrame, with annotations and a color map
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f')

# Set the title, x-axis label, and y-axis label for the heatmap
plt.title('Heatmap of Transaction Amounts by Day and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')

# Display the heatmap
plt.show()
```
