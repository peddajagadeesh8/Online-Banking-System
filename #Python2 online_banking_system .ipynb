```python
# Query data from the 'Customers' table and count the number of customers in each age group
cursor.execute('SELECT CASE \
                        WHEN Age >= 18 AND Age <= 25 THEN "18-25" \
                        WHEN Age >= 26 AND Age <= 35 THEN "26-35" \
                        WHEN Age >= 36 AND Age <= 45 THEN "36-45" \
                        WHEN Age >= 46 AND Age <= 55 THEN "46-55" \
                        ELSE "56+" \
                    END AS AgeGroup, \
                    COUNT(*) AS CustomerCount \
                FROM Customers \
                GROUP BY AgeGroup;')

# Fetch the data and store it into a Pandas DataFrame
customerCount_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

# Plot a horizontal bar chart of the number of customers per age group
import matplotlib.pyplot as plt
plt.figure(figsize=(9, 6))
plt.barh(customerCount_data['AgeGroup'], customerCount_data['CustomerCount'], color='skyblue')
plt.xlabel('Customer Count')
plt.ylabel('Age Group')
plt.title('Number of Customers per Age Group')
plt.tight_layout()
plt.show()
```

```python
# Assuming 'TransactionDate' is a datetime column
transactions_df['Day'] = transactions_df['TransactionDate'].dt.day_name()
transactions_df['Hour'] = transactions_df['TransactionDate'].dt.hour

# Pivot table for heatmap
heatmap_data = transactions_df.pivot_table(index='Day', columns='Hour', values='Amount', aggfunc='sum')

# Set the figure size for the heatmap
plt.figure(figsize=(12, 8))

# Create a heatmap using the heatmap_data DataFrame, with annotations and a color map
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f')

# Set the title, x-axis label, and y-axis label for the heatmap
plt.title('Heatmap of Transaction Amounts by Day and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')

# Display the heatmap
plt.show()
```

This code creates a heatmap showing the total transaction amounts by day of the week and hour of the day, using a pivot table and Seaborn's heatmap function, and annotating the heatmap with the transaction amounts.
